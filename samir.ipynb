```python
import matplotlib.pyplot as plt

data=[1,2,3,4,5]
plt.bar(range(len(data)),data,color='b')
plt.plot(data,marker='o',linestyle=':',color='r',linewidth=2)
plt.title('New Horizon',fontsize=40)
plt.xlabel('Data Science',fontsize=30)
plt.xlabel('Management',fontsize=20)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

price = [10, 20, 30, 40, 50]
demand = [100, 80, 60, 40, 20]
supply = [20, 40, 60, 80, 100]

plt.figure(figsize=(8,6))

plt.plot(price, demand, label='Demand', marker='o', color='red')
plt.plot(price, supply, label='Supply', marker='o', color='green')

plt.title('Supply and Demand Curve')
plt.xlabel('Price')
plt.ylabel('Quantity')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

data=[1,2,3,4,5]

plt.figure(figsize=(40, 10))
plt.pie(data, labels=data, autopct='%1.3f%%')
plt.title('Pie Chart')
plt.show()
```
